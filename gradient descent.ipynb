# 梯度下降

```python
import numpy as np

x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]

w = 1.0

cost_list = []
epoch_list = []

def forward(x):
    return x * w


def cost(xs, ys):
    cost = 0
    for x, y in zip(xs, ys):
        y_pred = forward(x)
        cost += (y_pred - y)**2
    return cost / len(xs)


def gradient(xs, ys):
    grad = 0
    for x, y in zip(xs, ys):
        grad += 2 * x * (x * w - y)
    return grad / len(xs)
```

```python
import matplotlib.pyplot as plt
print('predict (before training)', 4, forward(4))

for epoch in range(100):
    cost_val = cost(x_data,y_data)
    grad_val = gradient(x_data,y_data)
    w -= 0.01*grad_val
    cost_list.append(cost_val)
    epoch_list.append(epoch)
    print('Epoch:', epoch, 'w = ',w,'loss = ',cost_val)
    
plt.plot(epoch_list, cost_list)
plt.ylabel('cost')
plt.xlabel('epoch')
plt.show()
```

# Stochastic Gradient Descent

```python
# Stochastic Gradient Descent
import numpy as np

x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]

w = 1.0

cost_list = []
epoch_list = []


def forward(x):
    return x * w


def loss(x, y):
    y_pred = forward(x)
    return (y_pred - y)**2


def gradient(xs, ys):
    return 2 * x * (x * w - y)
```

```python
import matplotlib.pyplot as plt

print('predict (before training)', 4, forward(4))

for epoch in range(100):
    for x, y in zip(x_data, y_data):
        grad = gradient(x, y)
        w = w - 0.01 * grad
        print("\tgrad: ", x, y, grad)
        l = loss(x, y)
    print('Epoch:', epoch, 'w = ', w, 'loss = ', l)
```
